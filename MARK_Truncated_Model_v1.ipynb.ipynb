```python
import pandas as pd

file_path = "MARK Survey Data_Mother Sheet_17th July 2018_pharmEDGE_1.xlsx"

df = pd.read_excel(file_path, header=1)

df = df.dropna(axis=1, how="all")
df.columns = df.columns.str.strip()

df.head()
```

```python
numeric_cols = [
    "Age\n(years)",
    "Total cholesterol\n(mg/dL)",
    "HDL-cholesterol\n(mg/dL)",
    "Systolic blood pressure\n(mmHg)",
    "BMI (kg/m2)",
    "10-years QRISK3 score (%)"
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
df["High_Risk"] = (df["10-years QRISK3 score (%)"] >= 20).astype(int)

df["High_Risk"].value_counts()
```

```python
features = [
    "Age\n(years)",
    "Gender",
    "BMI (kg/m2)",
    "Smoking status",
    "Systolic blood pressure\n(mmHg)",
    "Blood pressure treatment",
    "Angina or heart attack",
    "Total cholesterol\n(mg/dL)",
    "HDL-cholesterol\n(mg/dL)",
    "Chronic kidney disease"
]

X = df[features].copy()
y = df["High_Risk"]
```

```python
X.columns = [
    "Age",
    "Sex",
    "BMI",
    "Smoking",
    "SystolicBP",
    "BPTreatment",
    "Angina",
    "TotalCholesterol",
    "HDL",
    "CKD"
]
```

```python
X["CholHDLRatio"] = X["TotalCholesterol"] / X["HDL"]
X = X.drop(columns=["HDL"])
```

```python
X.columns
```

```python
X = X.apply(pd.to_numeric, errors="coerce")
```

```python
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

model = CatBoostClassifier(
    iterations=300,
    depth=6,
    learning_rate=0.05,
    verbose=0
)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
import pickle

pickle.dump(model, open("truncated_model_v1.pkl", "wb"))
```
